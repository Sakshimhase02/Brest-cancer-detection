# importng labraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_1.csv')
```

```python
df
```

# Data cleaning

```python
df.drop(df.columns[[-1 , 0]], axis = 1 , inplace = True)
```

```python
df
```

```python
df['diagnosis'].value_counts()  #check the how many are breast cancer affectet (B , M)
```

# feature scaling

```python
from sklearn.model_selection import train_test_split
```

```python
diag_map = {'M':1,'B':0} 
```

```python
df['diagnosis'] = df['diagnosis'].map(diag_map)
```

```python
df
```

```python
x = df[['radius_mean' , 'perimeter_mean' , 'area_mean' , 'concavity_mean' ,'concave points_mean']] #randomly select 5 features modeling purpose if you want change you can change also
```

```python
x
```

```python
y = df['diagnosis']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

# model - KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train , y_train)
```

```python
knn_y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(knn_y_pred , y_test)
```

# model - logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr_y_pred = lr.predict(x_test)
```

```python
accuracy_score(lr_y_pred , y_test)
```

# model - Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train , y_train)
```

```python
gnb_y_pred = gnb.predict(x_test)
```

```python
accuracy_score(gnb_y_pred , y_test)
```

# model - k fold cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_all = []
```

```python
cvs_all = []
```

# 1 . KNN MODEL FIRST TAKE 

```python
scores = cross_val_score(knn , x ,y , cv = 10)    #10 time shuffled data

accuracy_all.append(accuracy_score(knn_y_pred , y_test))

cvs_all.append(np.mean(scores))
print("Accuracy : {0:2%}".format(accuracy_score(knn_y_pred , y_test)))

print("Cross Validation Score: {0:.2%} (+ /-  {1:.2%})".format(np.mean(scores),np.std(scores)*2))
```
